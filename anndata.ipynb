# Introduction to AnnData (Rewritten)

This notebook explains the fundamentals of the AnnData structure, a core data container used in single-cell RNA sequencing analysis. It is designed to efficiently store matrix-like biological data along with associated metadata.

## What is AnnData?

AnnData is used to manage datasets where rows represent observations (e.g., cells) and columns represent variables (e.g., genes). It supports sparse data, metadata storage, and structured annotations, making it highly suitable for scRNA-seq workflows.

```python
import numpy as np
import pandas as pd
import anndata as ad
from scipy.sparse import csr_matrix
```

## Creating an AnnData Object

We generate a random sparse matrix to simulate gene expression data.

```python
counts = csr_matrix(np.random.poisson(1, size=(100, 2000)), dtype=np.float32)
adata = ad.AnnData(counts)
adata
```

## Assigning Identifiers

Each observation (cell) and variable (gene) is labeled for easier indexing.

```python
adata.obs_names = [f'Cell_{i}' for i in range(adata.n_obs)]
adata.var_names = [f'Gene_{i}' for i in range(adata.n_vars)]
```

## Subsetting Data

AnnData allows slicing similar to pandas DataFrames.

```python
adata[['Cell_1','Cell_10'], ['Gene_5','Gene_100']]
```

## Adding Metadata

Metadata for cells is stored in `.obs` and for genes in `.var`.

```python
cell_types = np.random.choice(['B','T','Monocyte'], size=adata.n_obs)
adata.obs['cell_type'] = pd.Categorical(cell_types)
adata.obs.head()
```

## Filtering Using Metadata

We can subset cells based on annotations.

```python
adata[adata.obs['cell_type'] == 'B']
```

## Multi-dimensional Metadata

Use `.obsm` and `.varm` to store matrices such as embeddings.

```python
adata.obsm['X_umap'] = np.random.normal(0,1,(adata.n_obs,2))
adata.varm['gene_features'] = np.random.normal(0,1,(adata.n_vars,5))
```

## Unstructured Metadata

General information can be stored in `.uns`.

```python
adata.uns['info'] = {'example':[1,2,3]}
```

## Layers

Different versions of the dataset (e.g., transformed values) can be stored in layers.

```python
adata.layers['log_data'] = np.log1p(adata.X)
```

## Convert to DataFrame

AnnData can output a pandas DataFrame.

```python
adata.to_df().head()
```

## Saving Data

AnnData supports saving to .h5ad format.

```python
adata.write('results.h5ad')
```

## Summary

This notebook covered the key components of AnnData, including data storage, metadata handling, slicing, and saving results. It forms the foundation for more advanced single-cell analysis workflows.
